Suppose you roll 3 fair 100 sided die, what is the EV of the lowest roll?

```python
import random

def roll_dice(): #each time we call this function, it will return a random integer between 1 and 100, simulating the roll of a 100-sided die.
    return random.randint(1, 100)

minval = [] #Initialise an empty list to store the minimum values from each iteration of rolling three dice.
num_iters = 10000
for i in range(num_iters): #for each iteration 

    d1 = roll_dice() #Roll the first die and store the result in d1.
    d2 = roll_dice() #Roll the second die ...
    d3 = roll_dice()

    minval.append(min(d1, d2, d3)) 

print(sum(minval) / num_iters)
```

Suppose you roll 3 fair 100 sided die, what is the EV of highest roll?

```python
import random
def roll_dice(): 
    return random.randint(1, 100)

maxval = [] 
num_iters = 10000
for i in range(num_iters):  

    d1 = roll_dice() 
    d2 = roll_dice() 
    d3 = roll_dice()

    maxval.append(max(d1, d2, d3)) 

print(sum(maxval) / num_iters)
```
