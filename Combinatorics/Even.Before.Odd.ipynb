Suppose you roll a fair 6-sided die until you've seen ALL 6 faces. What is the probability you won't see an odd numbered face until you have seen all even numbered faces?

1. Firstly, there are 6! total permutations of the die faces.
2. We want situations where we have ONLY seen even numbers as the first 3 sightings.
3. So [Even, Even, Even, Odd, Odd, Odd] is the ONLY appropriate arrangement.
4. We can have 3! ways to arrange our even set. There are 3! ways to arrange the odd set.
5. Multiply these probabilities (each of the 6 arrangements of even numbers can correspond to 6 possible arrangements of our odd numbers, so 6x6)
6. Answer is 1/20 (0.05)

```python
import random

def roll_dice():
    return random.randint(1, 6)

valid_cases = 0 
num_trials = 10000

for i in range(num_trials):
    seen_faces = set() #Creates a set to store which die faces have been seen so far, automatically keep unique values
    while(True): #Infinite loop for repeated rolling within a single trial until a stopping condition is met. True is always true, so the loop will continue indefinitely until we break out of it.

        roll = roll_dice() #For each trial, we call our roll_dice function to simulate rolling a die and store the result in the variable roll.
        seen_faces.add(roll) #Inserts our new roll into the seen_faces set, which will automatically handle duplicates and only keep unique die faces.

        if roll % 2 == 1 and len(seen_faces) < 4: #KEY STOPPING CRITERION
            break
        if len(seen_faces) == 6:
            valid_cases += 1
            break


print(valid_cases / num_trials)
```

roll % 2 is MODULO operator: 4 % 2 = 0, 5 % 2 = 1, 6 % 2 = 0, 1 % 2 = 1, (used to check odd/even)

So the first stopping condition is if roll value is ODD and there are 3 or less seen faces then BREAK. This violates our VALID CASES, where we must roll 3 even numbered faces BEFORE rolling any odd numbered faces. 

If we have seen all possible dice roll faces, and we haven't already broken, this is now a valid sequence, so add to valid cases.
